```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


from pachner_traversal.utils import data_path, set_style

set_style()
```

```python
result_path = data_path / "results" / "direct_ascent" / "determinant_alexander_polynomial" / "20250923_2115"
assert result_path.exists()
```

```python
all_knotteds = pd.read_csv(result_path / 'all_knotteds.csv')
avg_scores = pd.read_csv(result_path / 'avg_scores.csv')
count_unknotteds = pd.read_csv(result_path / 'count_unknotteds.csv')
isos = pd.read_csv(result_path / 'isos.csv')
p_knotteds = pd.read_csv(result_path / 'p_knotteds.csv')
scores = pd.read_csv(result_path / 'scores.csv')
betas = pd.read_csv(result_path / 'betas.csv')
```

```python
fig, ax = plt.subplots()

ax.plot(scores[betas['beta'] == 10].max(), color='red', label=r'$\beta=10$', marker='o')
ax.plot(scores[betas['beta'] == np.sqrt(10)].max(), color='orange', label=r'$\beta=\sqrt{10}$')
ax.plot(scores[betas['beta'] == 1].max(), color='green', label=r'$\beta=1$')
ax.plot(scores[(betas['beta'] > 0.1) & (betas['beta'] < 1)].max(), color='purple', label=r'$\beta = \sqrt{0.1}$')
ax.plot(scores[betas['beta'] == 0.1].max(), color='blue', label=r'$\beta=0.1$')
ax.legend()
plt.show()
```
