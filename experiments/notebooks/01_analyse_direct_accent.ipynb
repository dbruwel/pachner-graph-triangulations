```python
fig, ax = plt.subplots()

ax.plot(np.arange(2, 30), scores[betas['beta'] == 10].max(), color='red', label=r'$\beta=10$')
ax.plot(np.arange(2, 30), scores[betas['beta'] == np.sqrt(10)].max(), color='orange', label=r'$\beta=\sqrt{10}$', ls='--')
ax.plot(np.arange(2, 30), scores[betas['beta'] == 1].max(), color='green', label=r'$\beta=1$')
ax.plot(np.arange(2, 30), scores[(betas['beta'] > 0.1) & (betas['beta'] < 1)].max(), color='purple', label=r'$\beta = \sqrt{0.1}$', ls='--')
ax.plot(np.arange(2, 30), scores[betas['beta'] == 0.1].max(), color='blue', label=r'$\beta=0.1$')
ax.set_xlabel('Number of Tetrahedra')
ax.set_ylabel('Max score')
ax.legend()
plt.show()
```
