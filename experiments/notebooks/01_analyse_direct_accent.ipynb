```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pathlib


from pachner_traversal.utils import data_path, set_style

set_style()
```

```python
result_path = data_path / "results" / "direct_ascent" / "norm_alexander_polynomial" / "20250921_1323"
```

```python
all_knotteds = pd.read_csv(result_path / 'all_knotteds.csv')
avg_scores = pd.read_csv(result_path / 'avg_scores.csv')
count_unknotteds = pd.read_csv(result_path / 'count_unknotteds.csv')
isos = pd.read_csv(result_path / 'isos.csv')
p_knotteds = pd.read_csv(result_path / 'p_knotteds.csv')
scores = pd.read_csv(result_path / 'scores.csv')
betas = pd.read_csv(result_path / 'betas.csv')
```

```python
plt.plot(scores[betas['beta'] == 10].mean(), color='red')
plt.plot(scores[betas['beta'] == np.sqrt(10)].mean(), color='orange')
plt.plot(scores[betas['beta'] == 1].mean(), color='green')
plt.plot(scores[(betas['beta'] > 0.1) & (betas['beta'] < 1)].mean(), color='purple')
plt.plot(scores[betas['beta'] == 0.1].mean(), color='blue')
plt.show()
```
