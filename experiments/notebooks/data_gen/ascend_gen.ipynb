```python
import h5py
from pachner_traversal.mcmc import neighbours
from regina import Triangulation3
import random
```

```python
root_path = "/home/dbruwel/main/honours/pachner_graph_triangulations/data/input_data/dehydration/processed/"
```

```python
with h5py.File(
    root_path + "spheres_10_scored.hdf5",
    "r",
) as f:
    raw_data = f["isos"][:]  # type: ignore
    data10 = [d.decode("utf-8") for d in raw_data]  # type: ignore

random.shuffle(data10)
```

```python
def step_up_data(data):
    up_data = set()
    p_count = 0
    for sig in data:
        f_vec = Triangulation3.rehydrate(sig).fVector()
        nbrs = neighbours(sig, f_vec, 1).keys()
        up_data.update(nbrs)

        if len(up_data) >= len(data):
            return up_data
        else:
            p_count += 1
            if p_count % 1_000 == 0:
                print(len(up_data))

    raise RuntimeError("Not enough samples.")
```

```python
def save_hdf5(file_path, data, dataset_name='isos'):
    if not data:
        raise ValueError("The provided string list is empty.")

    string_length = len(data[0].encode('utf-8'))

    string_dt = h5py.string_dtype(encoding='utf-8', length=string_length)

    with h5py.File(file_path, 'w') as f:
        dset = f.create_dataset(
            dataset_name,
            shape=(len(data),),
            dtype=string_dt
        )

        dset[:] = data
```

```python
print("data10")
data10 = data10.copy()
random.shuffle(data10)
save_hdf5(root_path + "spheres_10.hdf5", data10)
print("data11")
data11 = list(step_up_data(data10))[:4_000_000]
random.shuffle(data11)
save_hdf5(root_path + "spheres_11.hdf5", data11)
print("data12")
data12 = list(step_up_data(data11))[:4_000_000]
random.shuffle(data12)
save_hdf5(root_path + "spheres_12.hdf5", data12)
print("data13")
data13 = list(step_up_data(data12))[:4_000_000]
random.shuffle(data13)
save_hdf5(root_path + "spheres_13.hdf5", data13)
print("data14")
data14 = list(step_up_data(data13))[:4_000_000]
random.shuffle(data14)
save_hdf5(root_path + "spheres_14.hdf5", data14)
print("data15")
data15 = list(step_up_data(data14))[:4_000_000]
random.shuffle(data15)
save_hdf5(root_path + "spheres_15.hdf5", data15)
print("data16")
data16 = list(step_up_data(data15))[:4_000_000]
random.shuffle(data16)
save_hdf5(root_path + "spheres_16.hdf5", data16)
print("data17")
data17 = list(step_up_data(data16))[:4_000_000]
random.shuffle(data17)
save_hdf5(root_path + "spheres_17.hdf5", data17)
print("data18")
data18 = list(step_up_data(data17))[:4_000_000]
random.shuffle(data18)
save_hdf5(root_path + "spheres_18.hdf5", data18)
print("data19")
data19 = list(step_up_data(data18))[:4_000_000]
random.shuffle(data19)
save_hdf5(root_path + "spheres_19.hdf5", data19)
print("data20")
data20 = list(step_up_data(data19))[:4_000_000]
random.shuffle(data20)
save_hdf5(root_path + "spheres_20.hdf5", data20)
```
