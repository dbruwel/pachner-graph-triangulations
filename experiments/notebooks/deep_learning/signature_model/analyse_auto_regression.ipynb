```python
import os
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import regina
from pachner_traversal.utils import set_style

set_style()
```

```python
file_path = "/home/dbruwel/main/honours/pachner_graph_triangulations/data/results/sgd_models_dehydration/autoregression_ntet/autoregression_10_0.001/generated_samples.txt"

samp_path = file_path

with open(samp_path, "r") as f:
    lines = f.readlines()

samps_str = [line.strip() for line in lines]
```

```python
def check_triangulation(iso_sig):
    try:
        T = regina.engine.Triangulation3.rehydrate(iso_sig)
        loads = True
        valid = T.isValid()
        closed = T.isClosed()
        sphere = T.fundamentalGroup().recogniseGroup() == '0'
    except Exception:
        loads = False
        valid = False
        closed = False
        sphere = False

    return loads, valid, closed, sphere
```

```python
checks = []

for i, iso in enumerate(samps_str):
    checks.append(check_triangulation(iso))
loads = [c[0] for c in checks]
valid = [c[1] for c in checks]
closed = [c[2] for c in checks]
sphere = [c[3] for c in checks]
```

```python
p_loads = np.sum(loads) / len(loads)
p_valid = np.sum(valid) / len(valid)
p_closed = np.sum(closed) / len(closed)
p_sphere = np.sum(sphere) / len(sphere)

# print(f"Loads: {p_loads:.4f}")
# print(f"Valid: {p_valid:.4f}")
# print(f"Closed: {p_closed:.4f}")
# print(f"Sphere: {p_sphere:.4f}")

print(f"{p_loads:.4f}, {p_valid:.4f}, {p_closed:.4f}, {p_sphere:.4f}")
```
