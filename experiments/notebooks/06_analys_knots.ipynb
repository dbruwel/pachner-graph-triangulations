```python
import regina
from snappy import Manifold
import matplotlib.pyplot as plt
import numpy as np
```

```python
iso_sig = 'EMLLLvzPPzwPLMQzAPzAQabddglnjokkqnouqswstuvwzyzBCDCDgagbaaagacvaggaccavgfbcafcoaaco'
ids = {}
for edge_index in range(31):
    try:
        tri = regina.engine.Triangulation3.fromIsoSig(iso_sig)
        edge = tri.edge(edge_index)

        tri.pinchEdge(edge)

        ids[edge_index] = Manifold(tri).alexander_polynomial().degree()
    except:
        ids[edge_index] = 0
```

```python
np.array(list(ids.values())).argmax()
```
